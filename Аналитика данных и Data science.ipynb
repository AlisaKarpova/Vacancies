Москва Data Science

```python
import requests
from bs4 import BeautifulSoup

j = 0
m = 0
s = 0

base_url = 'https://career.habr.com/vacancies?locations[]=c_678&q=data%20science&type=all'

for i in range(1, 3):
    page_url = f'{base_url}&page={i}'
    response = requests.get(page_url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        for i in soup.find_all(class_ ="link-comp link-comp--appearance-dark"):
          if 'Junior' in i.text:
            j += 1
          if 'Middle' in i.text:
            m += 1
          if 'Senior' in i.text:
            s += 1
print(j)
print(m)
print(s)
```

```python
import matplotlib.pyplot as plt

categories = ['Junior', 'Middle', 'Senior']
counts = [j, m, s]

plt.bar(categories, counts)
plt.xlabel('Experience Level')
plt.ylabel('Number of Vacancies')
plt.title('Data Science')
plt.show()
```

Москва Аналитика данных

```python
import requests
from bs4 import BeautifulSoup

j1 = 0
m1 = 0
s1 = 0

base_url = 'https://career.habr.com/vacancies?locations[]=c_678&q=аналитика%20данных&type=all'

for i in range(1, 10):
    page_url = f'{base_url}&page={i}'
    response = requests.get(page_url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        for i in soup.find_all(class_ ="link-comp link-comp--appearance-dark"):
          if 'Junior' in i.text:
            j1 += 1
          if 'Middle' in i.text:
            m1 += 1
          if 'Senior' in i.text:
            s1 += 1
print(j1)
print(m1)
print(s1)

```

```python
import matplotlib.pyplot as plt

categories = ['Junior', 'Middle', 'Senior']
counts = [j1, m1, s1]

plt.bar(categories, counts)
plt.xlabel('Experience Level')
plt.ylabel('Number of Vacancies')
plt.title('Аналитика данных')
plt.show()
```

Data science и аналитика данных в Москве

```python
import matplotlib.pyplot as plt

categories = ['Junior', 'Middle', 'Senior']
counts = [j, m, s]
additional_counts = [j1, m1, s1]
bar_width = 0.35

plt.bar(categories, counts, width=bar_width, label='Data science', align='center' )
plt.bar(categories, additional_counts, width=bar_width, label='Аналитика данных', align='edge')

plt.xlabel('Уровень квалификации')
plt.ylabel('Количество вакансий')
plt.title('Аналитика данных и Data science')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
table_data = [['', 'Аналитика данных', 'Data science', 'Всего вакансий'], ['Junior', j, j1, j + j1], ['Middle', m, m1, m + m1], ['Senior', s, s1, s + s1], ['Количество вакансий', j+m+s, j1+m1+s1, j + j1 + m + m1 + s + s1]]

table = ax.table(cellText=table_data, loc='center')

table.set_fontsize(14)
table.scale(1, 4)
ax.axis('off')

plt.show()
```
